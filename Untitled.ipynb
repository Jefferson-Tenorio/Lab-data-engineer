```python
import psycopg2
import time

# CONFIGURAÇÃO CORRETA PARA O SEU DOCKER
DB_CONFIG = {
    'dbname': 'lab_dados',
    'user': 'postgres',
    'password': 'root',
    'host': 'db',          # MUITO IMPORTANTE: dentro do Docker é "db"
    'port': '5432'
}

def run_wal_experiment():
    conn = None
    cur = None
    
    try:
        conn = psycopg2.connect(**DB_CONFIG)
        conn.autocommit = False
        cur = conn.cursor()
        
        # 1️⃣ Snapshot inicial do LSN
        cur.execute("SELECT pg_current_wal_lsn();")
        lsn_antigo = cur.fetchone()[0]
        print(f"📍 LSN INICIAL: {lsn_antigo}")
        
        # 2️⃣ Gerar carga
        print("⚡ Gerando carga (10.000 inserts)...")
        cur.execute("""
            CREATE TABLE IF NOT EXISTS lab_wal (
                id serial PRIMARY KEY,
                data text
            );
        """)
        
        payload = "x" * 100  # 100 bytes por linha
        
        for _ in range(10000):
            cur.execute(
                "INSERT INTO lab_wal (data) VALUES (%s)",
                (payload,)
            )
        
        conn.commit()
        print("✅ Commit realizado.")
        
        # 3️⃣ Snapshot final
        cur.execute("SELECT pg_current_wal_lsn();")
        lsn_novo = cur.fetchone()[0]
        print(f"📍 LSN FINAL: {lsn_novo}")
        
        # 4️⃣ Diferença real de WAL gerado
        cur.execute("SELECT pg_wal_lsn_diff(%s, %s);", (lsn_novo, lsn_antigo))
        diff_bytes = cur.fetchone()[0]
        
        print("-" * 40)
        print("📊 Resultado:")
        print(f"WAL gerado: {diff_bytes} bytes")
        print(f"WAL gerado: {diff_bytes / 1024 / 1024:.2f} MB")
        print("-" * 40)
        
    except Exception as e:
        print(f"❌ Erro: {e}")
        
    finally:
        if cur:
            cur.close()
        if conn:
            conn.close()

run_wal_experiment()
```

```python
import psycopg2
import time

DB_CONFIG = {
    'dbname': 'lab_dados',
    'user': 'postgres',
    'password': 'root',
    'host': 'db',          # IMPORTANTE: dentro do Docker é "db"
    'port': '5432'
}

def measure_inserts(sync_mode, iterations=1000):
    conn = None
    cur = None
    
    try:
        conn = psycopg2.connect(**DB_CONFIG)
        conn.autocommit = False
        cur = conn.cursor()
        
        # Configura synchronous_commit para a sessão atual
        cur.execute("SET synchronous_commit TO %s;", (sync_mode,))
        
        # Cria tabela se não existir
        cur.execute("""
            CREATE TABLE IF NOT EXISTS speed_test (
                id serial PRIMARY KEY,
                data text
            );
        """)
        
        # Limpa tabela para teste justo
        cur.execute("TRUNCATE speed_test;")
        conn.commit()
        
        start_time = time.time()
        
        for _ in range(iterations):
            cur.execute(
                "INSERT INTO speed_test (data) VALUES (%s);",
                ("teste de velocidade",)
            )
            conn.commit()  # força flush por linha
        
        end_time = time.time()
        total_time = end_time - start_time
        
        print(
            f"Modo: {sync_mode.upper()} | "
            f"Inserts: {iterations} | "
            f"Tempo Total: {total_time:.4f}s | "
            f"TPS: {iterations/total_time:.2f}"
        )
        
    except Exception as e:
        print(f"Erro no modo {sync_mode}: {e}")
        
    finally:
        if cur:
            cur.close()
        if conn:
            conn.close()


if __name__ == "__main__":
    print("--- INICIANDO BATTLE ROYALE DE I/O ---")
    
    # Modo Seguro (durável)
    measure_inserts('on', 1000)
    
    # Modo Assíncrono (menos durável)
    measure_inserts('off', 1000)
```
