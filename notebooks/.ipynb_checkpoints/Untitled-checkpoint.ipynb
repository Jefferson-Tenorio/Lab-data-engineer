```python
import pyspark
from pyspark.sql import SparkSession
from delta import *

# Configuração ROBUSTA (Copie e use sempre esta)
builder = pyspark.sql.SparkSession.builder.appName("RealEstateLakehouse") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    \
    # 1. Configurações de Conexão com MinIO (Obrigatórias para s3a:// funcionar)
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio:9000") \
    .config("spark.hadoop.fs.s3a.access.key", "minio_admin") \
    .config("spark.hadoop.fs.s3a.secret.key", "minio_password") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false") \
    \
    # 2. Injeção dos JARs (Obrigatório para corrigir o ClassNotFoundException)
    # hadoop-aws: Ensina o Spark a falar com S3
    # delta-spark: Ensina o Spark a falar com Delta Lake
    .config("spark.jars.packages", "io.delta:delta-spark_2.12:3.2.0,org.apache.hadoop:hadoop-aws:3.3.4")

# Cria a sessão
spark = configure_spark_with_delta_pip(builder).getOrCreate()

print("Spark configurado com S3A e Delta! Pode rodar o write agora.")
```

```python
from faker import Faker
import pandas as pd
import random

fake = Faker("pt_BR")

dados = []

for _ in range(1000):
    dados.append({
        "id": random.randint(1, 100000),
        "nome": fake.name(),
        "cidade": fake.city(),
        "preco": random.randint(100000, 1000000),
        "data_cadastro": fake.date_this_year()
    })

df_pandas = pd.DataFrame(dados)

df_pandas.head()
```

```python
df_spark = spark.createDataFrame(df_pandas)
# Agora vai funcionar
df_spark.write.mode("overwrite").parquet("s3a://lakehouse/landing/imoveis/")
df_spark.printSchema()
df_spark.show(5)
```
